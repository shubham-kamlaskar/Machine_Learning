```python
cred_obj['Name'].fillna(cred_obj['Name'].mode()[0],inplace=True)
cred_obj['Type_of_Loan'].fillna(cred_obj['Type_of_Loan'].mode()[0],inplace=True)
cred_obj['Num_of_Delayed_Payment'].fillna(cred_obj['Num_of_Delayed_Payment'].mode()[0],inplace=True)
```

```python
cred_int = cred.select_dtypes(include=int)
cred_int.info()
cred_int
```

```python
cred_float = cred.select_dtypes(include=float).round(2)
cred_float.info()
```

```python
cred_float['Monthly_Inhand_Salary'] = cred_float['Monthly_Inhand_Salary'].fillna(cred_float['Monthly_Inhand_Salary'].mean()).round(2)
cred_float['Num_Credit_Inquiries'] = cred_float['Num_Credit_Inquiries'].fillna(cred_float['Num_Credit_Inquiries'].mean())
cred_float.info()
```

```python
print(cred_int.shape)
print(cred_float.shape)
```

```python
cred_num = pd.concat([cred_int,cred_float],axis=1)
cred_num['Monthly_Inhand_Salary']=cred_num['Monthly_Inhand_Salary'].round(2)
cred_num
```

```python
import re

# cred_obj['Annual_Income'] = re.sub('_','',cred_obj['Annual_Income'])

cred_obj['Annual_Income'] = cred_obj['Annual_Income'].apply(lambda x : re.sub('_','',x))
```

```python
cred_obj['Annual_Income']= cred_obj['Annual_Income'].astype(float).round(2)
```

```python
cred_obj['Num_of_Delayed_Payment'] = cred_obj['Num_of_Delayed_Payment'].apply(lambda x : re.sub('[0-9]+_','nan',x))
```
