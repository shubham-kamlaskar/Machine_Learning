```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
```

```python
death = pd.read_csv('annual_deaths_by_causes.csv')
death
```

```python
death.shape
```

```python
death.info()
```

```python
death.isnull().sum().sort_values(ascending=False)
```

```python
death['terrorism'].fillna(0,inplace=True)
```

```python
death.isnull().sum().sort_values(ascending=False)
```

```python
death.drop(['code'],axis=1,inplace=True)
```

```python
if death['malaria'].isnull().sum()==0:
    print('null')
else:
    print('non null')
```

```python
null_col = []
for i in death.columns:
    if death[i].isnull().sum()!=0:
        null_col.append(i)

len(null_col)
```

```python
for i in null_col:
    print(i,':',death[i].mean().round(2))
    death[i] = death[i].fillna(death[i].mean().round(2))
```

```python
death.info()
```

for i in null_col:
    plt.subplot(len(null_col),1,len(i))
    sns.displot(death[i])
